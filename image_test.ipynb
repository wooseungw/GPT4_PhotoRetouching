```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from edit_img import *
processor = ImageProcessor("C:/Workspace/gpt4_picturecorrection/img_dir/lena.png")

images = []
titles = []

#원본 이미지
image = processor.origin_img
images.append(image)
titles.append('Original')

# 밝기 조정
image = processor.adjust_brightness(0.5)
images.append(image)
titles.append('Brightness adjusted')

# 대비 조정
image = processor.adjust_contrast(1.5)
images.append(image)
titles.append('Contrast adjusted')

# 노이즈 감소
image = processor.reduce_noise(1)
images.append(image)
titles.append('Noise reduced')

# 색상 조정
image = processor.adjust_hue(1.5)
images.append(image)
titles.append('Hue adjusted')

# 채도 조정
image = processor.adjust_saturation(1.5)
images.append(image)
titles.append('Saturation adjusted')

# 명도 조정
image = processor.adjust_value(1.5)
images.append(image)
titles.append('Value adjusted')

# 선명도 조정
image = processor.sharpen(1.5)
images.append(image)
titles.append('Sharpened')

# 블러 효과
image = processor.blur(0.5)
images.append(image)
titles.append('Blurred')

# 이미지들을 하나의 그림으로 표시
fig = plt.figure(figsize=(10, 10))
for i in range(len(images)):
    ax = fig.add_subplot(3, 3, i+1)
    ax.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))  # OpenCV는 BGR로 이미지를 처리하므로 RGB로 변환
    ax.set_title(titles[i])
    ax.axis('off')

plt.show()
```
